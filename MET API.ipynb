```python
## Review the different museum departments
response = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/departments')
json_data = response.json()
json_formatted_str = json.dumps(json_data, indent=2)
print(json_formatted_str)
```

```python
##Creates an empty Dataframe
columns = ['ObjectID', 'Department','Classification','Medium','Object_Name','Title','Date','BeginDate','EndDate']
df = pd.DataFrame(columns=columns)
```

```python
##Creates a list of Object IDs from a single Department
response = requests.get("https://collectionapi.metmuseum.org/public/collection/v1/objects?departmentIds=9")
json_data = response.json()
id_list = json_data['objectIDs']
print(len(id_list))
```

#### Begin Downloading object data. This step is broken down into segments that were completed over the course of several days. 

```python
## Segments the object list into a shorter list using a range from id_list
short_list = id_list[0:70000]
print(len(short_list))
```

```python
##Iterates through the short_list and adds the ObjectID, Department, Classification, and Medium to a dataframe
start_time = time.time()
root = 'https://collectionapi.metmuseum.org/public/collection/v1/objects/'
for y in short_list:
    id = str(y)
    response = requests.get(root+id)
    json_data = response.json()
    class_value = json_data['classification']
    objectid = json_data['objectID']
    department = json_data['department']
    medium = json_data['medium']
    object_name = json_data['objectName']
    title = json_data['title']
    date = json_data['objectDate']
    begin_date = json_data['objectBeginDate']
    end_date = json_data['objectEndDate']
    new_row = pd.DataFrame({'ObjectID': [objectid], 'Department': [department], 'Classification': [class_value], 'Medium': [medium], 'Object_Name':[object_name], 'Title':[title], 'Date': [date], 'BeginDate': [begin_date], 'EndDate': [end_date]})
    df = pd.concat([df, new_row], ignore_index=True)
end_time = time.time()
elapsed_time = end_time - start_time
print(elapsed_time)
```

#### At this point, return to the above step and run another segment.

Downloading Object Numbers separately because I forgot them last time. 
I'll join the Object Numbers to the rest of the data in sql. 

```python
##Creates an empty Dataframe
columns = ['ObjectID', 'ObjectNumber']
df = pd.DataFrame(columns=columns)
```

```python
response = requests.get("https://collectionapi.metmuseum.org/public/collection/v1/objects?")
json_data = response.json()
id_list = json_data['objectIDs']
print(len(id_list))
```

```python
##Creates a shorter list using a range from id_list
short_list = id_list[307035:488662]
print(len(short_list))
```

```python
start_time = time.time()
root = 'https://collectionapi.metmuseum.org/public/collection/v1/objects/'
for y in short_list:
    id = str(y)
    response = requests.get(root+id)
    json_data = response.json()
    objectid = json_data['objectID']
    objectnumber = json_data['accessionNumber']
    new_row = pd.DataFrame({'ObjectID': [objectid], 'ObjectNumber': [objectnumber], })
    df = pd.concat([df, new_row], ignore_index=True)
end_time = time.time()
elapsed_time = end_time - start_time
print(elapsed_time)
```
